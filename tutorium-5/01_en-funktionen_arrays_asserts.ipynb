## Functions in Python

Functions in Python are defined with the keyword `def`. This is followed by the function name and the parameters in parentheses.

```python
#Funktions-Definitionen
def division(x, y):
    return x / y
    
def hello():
    print("Hello")

#Funktionen aufrufen
print(division(1, 2))
hello()
```

## Lists in Python

Lists are one of the sequence data types in Python. They can contain elements of different data types, but are often used with only one data type.

```python
languages = ['haskell', 'python', 'java', 5.0]

print('First element: ' + languages[0])
print('Penultimate element: ' + languages[-2])

languages[3] = 'c'
languages.append('javascript')
print(languages)

print(languages.index('python'))
```

## Asserts in Python

Asserts can be used to check whether a certain condition is met. If this is not the case, an `AssertionError` will be triggered and the corresponding error message is printed out.

```python
def calc_average_grade(grades):
    ''' Calculates the grade point average from a list of grades.'''
    sum = 0    
    for grade in grades:
        assert grade >= 0 and grade <= 100, "Grades must be between 0 and 100"
        sum += grade    
    return sum / len(grades) 
        
calc_average_grade([99, 50, 75, 60])
```
