## Funktionen in Python

Funktionen in Python werden mit dem Schlüsselwort `def` definiert. Darauf folgt der Funktions-Name und die Parameter in Klammern.

```python
#Funktions-Definitionen
def division(x, y):
    return x / y
    
def hello():
    print("Hello")

#Funktionen aufrufen
print(division(1, 2))
hello()
```

## Listen in Python

Listen sind einer der Sequenz-Datentypen in Python. Sie können Elemente unterschiedlicher Datentypen enthalten, werden aber häufig mit nur einem Datentypen verwendet.

```python
languages = ['haskell', 'python', 'java', 5.0]

print('First element: ' + languages[0])
print('Penultimate element: ' + languages[-2])

languages[3] = 'c'
languages.append('javascript')
print(languages)

print(languages.index('python'))
```

## Asserts in Python

Mit Asserts lässt sich überprüfen, ob eine bestimmte Voraussetzung erfüllt ist. Trifft diese nicht zu, wird ein `AssertionError` ausgelöst und
die entsprechende Fehlermeldung ausgegeben.

```python
def calc_average_grade(grades):
    ''' Calculates the grade point average from a list of grades.'''
    sum = 0    
    for grade in grades:
        assert grade >= 0 and grade <= 100, "Grades must be between 0 and 100"
        sum += grade    
    return sum / len(grades) 
        
calc_average_grade([99, 50, 75, 60])
```
